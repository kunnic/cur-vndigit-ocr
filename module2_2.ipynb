```python
import json
import os

fileraw = 'result_0.json'
raw = os.path.join('data', fileraw)

with open(raw, 'r', encoding='utf-8') as f:
    data_sample = json.load(f)

print(data_sample) 
```

```python
def get_bounding_box_2_2(polygon):
    # Lấy danh sách tất cả tọa độ x và y từ 4 điểm
    x_coords = [point[0] for point in polygon]
    y_coords = [point[1] for point in polygon]

    # Tìm điểm bắt đầu (min) và tính kích thước (max - min)
    x = min(x_coords)
    y = min(y_coords)
    w = max(x_coords) - x
    h = max(y_coords) - y

    return {"x": x, "y": y, "w": w, "h": h}

for item in data_sample['texts']:
    text_content = item['text']
    poly_data = item['bounding_polygon']

    bbox = get_bounding_box_2_2(poly_data)

    print(f"Text: {text_content[:30]}... -> BBox chuẩn 2.2: {bbox}")
```
